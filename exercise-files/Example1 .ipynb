```python
# given a list of integers, sum the factorial of each integer 
def factorial(n):
    if n <= 1:
        return 1
    else:
        return n * factorial(n-1)
```

```python
factorial(5)
```

```python
factorial(4)
```

```python
factorial(3)
```

```python
numbers = range(1,6)
sum([factorial(one_number) for one_number in numbers])
```
