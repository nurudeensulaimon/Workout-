```python
def word_score(word):
    return sum([ord(letter) - 96
                for letter in word.lower()])
```

```python
word_score('abc')
```

```python
words = input("Enter some words here:")
```

```python
{one_word : word_score(one_word)
for one_word in words.split()}
```

```python
{one_word : word_score(one_word)
for one_word in words.split()
if len(one_word) >= 3 and len(one_word) <= 7}
```
