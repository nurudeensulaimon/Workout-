```python
def word_score(word):
    return sum([ord(letter) - 96
                for letter in word.lower()])
```

```python
def sums_to_n(n):
    return [(x,y)
            for x in range(0, n+1)
            for y in range(0, n+1)
            if x + y == n]
```

```python
sums_to_n(10)
```
