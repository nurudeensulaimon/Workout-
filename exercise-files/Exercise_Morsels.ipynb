```python
def with_previous(sequence):
    '''join a sequence with its previous element'''
    items = []
    for i in range(len(sequence)):
        if i == 0:
            items.append((sequence[i], None))
        else:
            items.append((sequence[i], sequence[i-1]))
    return items 
```

```python
with_previous('hello')
```

```python
def with_previous1(iterable):
    """previous each iterable item with item before it."""
    previous = None 
    items = []
    for item in iterable:
        items.append((item,previous))
        previous = item 
    return items
```

```python
with_previous1('hello')
```

```python
def with_previous(sequence):
    """provide each sequence item with item before it"""
    items = []
    for item, previous in zip(sequence,[None] + list(sequence)):
        items.append((item, previous))
    return items
```

```python
with_previous('hello')
```

```python
def with_previous(sequence):
    """provide each sequence item with item before it"""
    return zip(sequence,[None, *sequence])
```

```python
with_previous('hello')
```

```python
def with_previous(iterable, *, fillvalue=None):
    """Provide each iterable item with item before it"""
    previous = fillvalue 
    for item in iterable:
        yield item, previous
        previous = item 
        
```

```python
with_previous([1,1,2,3,4])
```
